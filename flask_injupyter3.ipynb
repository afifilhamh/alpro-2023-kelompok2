```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder

# Load your dataset
data_path = "HREmployee.csv"
df = pd.read_csv(data_path)
education_mapping = {1: "High School",
                    2: "College",
                    3: "Bachelor",
                    4: "Master",
                    5: "PhD"}

# Mengganti nilai dalam kolom 'Education' berdasarkan mapping yang didapat dari informasi dataset di Kaggle
df['Education'] = df['Education'].map(education_mapping)
df_asli = df.copy()
selected_columns = ['Age', 'Over18', 'Gender', 'MaritalStatus', 'Education',
                    'EducationField', 'NumCompaniesWorked', 'TotalWorkingYears',
                    'Department', 'MonthlyIncome', 'BusinessTravel',
                    'OverTime', 'JobRole']
df_asli = df_asli.loc[:, selected_columns]

df_asli
```

```python
from sklearn.model_selection import train_test_split


X = df_asli.drop(['JobRole'], axis=1)
y = df_asli['JobRole']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Initialize LabelEncoders for features and target
label_encoders = {}

# Encode categorical columns in the dataframe
for column in X_train.select_dtypes(include=['object']).columns:
    label_encoders[column] = LabelEncoder()
    X_train[column] = label_encoders[column].fit_transform(X_train[column])

for column in X_test.select_dtypes(include=['object']).columns:
    label_encoders[column] = LabelEncoder()
    X_test[column] = label_encoders[column].fit_transform(X_test[column])
```

```python
X_train
```

```python
X_test
```

```python
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

# Predictions on the test set
predictions = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, predictions)
accuracy
```
